```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the spotify Disease dataset
df = pd.read_csv('Mobile_Price.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Display basic statistics of the dataset
print(df.describe())

# Calculate and display major moments for numerical features
moments = df.describe().transpose()[['mean', '50%', 'std']]
skewness = df.skew()
kurtosis = df.kurtosis()

# Combine moments, skewness, and kurtosis into a single DataFrame
statistics_df = pd.concat([moments, skewness, kurtosis], axis=1)
statistics_df.columns = ['Mean', 'Median', 'Standard Deviation', 'Skewness', 'Kurtosis']
print()
# Display the statistics
print("Descriptive Statistics")
print(statistics_df)
```

```python
# Compute correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
#Categorical Graph
# Plot a pie chart for a categorical variable 
plt.figure(figsize=(8, 8))
labels = df['int_memory'].value_counts().index
sizes = df['int_memory'].value_counts().values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('price_range')
plt.axis('equal')
plt.show()
```

```python
# Create scatter plot
plt.scatter(df['battery_power'], df['ram'], color='orange')
plt.title('Relationship between Battery and Ram')
plt.xlabel('Battery')
plt.ylabel('Ram')
plt.show()
```

```python
#Statistical graph
# Plot correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
